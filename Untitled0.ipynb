```python
import tensorflow as tf
import pandas as pd
import os
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess
from tensorflow.keras.applications.mobilenet import preprocess_input as mobilenet_preprocess
from tensorflow.keras.applications import ResNet50, MobileNet
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization, Conv2D, Multiply
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve, auc, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import zipfile
import os

with zipfile.ZipFile("archive.zip", 'r') as zip_ref:
    zip_ref.extractall("archive")  # will create archive/train-image/image/ and archive/train-metadata.csv

```

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32
```

```python
image_dir = "archive/train-image/image/"
csv_path = "archive/train-metadata.csv"


df = pd.read_csv(csv_path)
df['image_path'] = df['isic_id'].apply(lambda x: os.path.join(image_dir, x + '.jpg'))
df = df[df['image_path'].apply(os.path.exists)]
print(f" Total valid images: {len(df)}")


def preprocess_train_resnet(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = resnet_preprocess(image)
    return image, label

def preprocess_val_resnet(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = resnet_preprocess(image)
    return image, label

def preprocess_train_mobilenet(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = mobilenet_preprocess(image)
    return image, label

def preprocess_val_mobilenet(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = mobilenet_preprocess(image)
    return image, label


image_paths = df['image_path'].values
labels = df['target'].values
dataset = tf.data.Dataset.from_tensor_slices((image_paths, labels))

train_size = int(0.8 * len(df))
train_dataset_resnet = dataset.take(train_size).map(preprocess_train_resnet).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_dataset_resnet = dataset.skip(train_size).map(preprocess_val_resnet).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

train_dataset_mobilenet = dataset.take(train_size).map(preprocess_train_mobilenet).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_dataset_mobilenet = dataset.skip(train_size).map(preprocess_val_mobilenet).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
```

```python
base_resnet = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_resnet.trainable = False
x = base_resnet.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
resnet_output = Dense(1, activation='sigmoid')(x)

resnet_model = Model(inputs=base_resnet.input, outputs=resnet_output)
resnet_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])


base_mobilenet = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_mobilenet.trainable = False
y = base_mobilenet.output
y = GlobalAveragePooling2D()(y)
y = Dense(128, activation='relu')(y)
mobilenet_output = Dense(1, activation='sigmoid')(y)

mobilenet_model = Model(inputs=base_mobilenet.input, outputs=mobilenet_output)
mobilenet_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
```

```python
resnet_history = resnet_model.fit(
    train_dataset_resnet,
    validation_data=val_dataset_resnet,
    epochs=5
)
```

```python

mobilenet_history = mobilenet_model.fit(
    train_dataset_mobilenet,
    validation_data=val_dataset_mobilenet,
    epochs=5
)
```

```python

print("ResNet-50 Training Results:")
print(f"Final Train Accuracy: {resnet_history.history['accuracy'][-1]:.4f}")
print(f"Final Val Accuracy: {resnet_history.history['val_accuracy'][-1]:.4f}")
print(f"Final Train Loss: {resnet_history.history['loss'][-1]:.4f}")
print(f"Final Val Loss: {resnet_history.history['val_loss'][-1]:.4f}")

print("\nMobileNet Training Results:")
print(f"Final Train Accuracy: {mobilenet_history.history['accuracy'][-1]:.4f}")
print(f"Final Val Accuracy: {mobilenet_history.history['val_accuracy'][-1]:.4f}")
print(f"Final Train Loss: {mobilenet_history.history['loss'][-1]:.4f}")
print(f"Final Val Loss: {mobilenet_history.history['val_loss'][-1]:.4f}")


import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))


plt.subplot(1, 2, 1)
plt.plot(resnet_history.history['accuracy'], label='ResNet Train Acc')
plt.plot(resnet_history.history['val_accuracy'], label='ResNet Val Acc')
plt.plot(mobilenet_history.history['accuracy'], label='MobileNet Train Acc')
plt.plot(mobilenet_history.history['val_accuracy'], label='MobileNet Val Acc')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(resnet_history.history['loss'], label='ResNet Train Loss')
plt.plot(resnet_history.history['val_loss'], label='ResNet Val Loss')
plt.plot(mobilenet_history.history['loss'], label='MobileNet Train Loss')
plt.plot(mobilenet_history.history['val_loss'], label='MobileNet Val Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.applications import ResNet50, MobileNet
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping

# Model creation functions (unchanged)
def create_resnet_model():
    base_resnet = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_resnet.trainable = False
    x = base_resnet.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)
    return Model(inputs=base_resnet.input, outputs=output)



# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Tune ResNet-50
resnet_model = create_resnet_model()
resnet_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                     loss='binary_crossentropy', metrics=['accuracy', 'Precision', 'Recall'])
print("\nTuning ResNet-50 (lr=0.0001, epochs=10)...")
resnet_history_tuned = resnet_model.fit(train_dataset_resnet, validation_data=val_dataset_resnet,
                                        epochs=10, callbacks=[early_stopping])
```

```python
# Check results
print("\nResNet-50 Tuned Results:")
print(f"Val Accuracy: {max(resnet_history_tuned.history['val_accuracy']):.4f}")
print(f"Val Loss: {min(resnet_history_tuned.history['val_loss']):.4f}")
```

```python
print(f"Val Accuracy: {max(resnet_history_tuned.history['val_accuracy']):.4f}")
```

```python
def create_mobilenet_model():
    base_mobilenet = MobileNet(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_mobilenet.trainable = False
    x = base_mobilenet.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)
    return Model(inputs=base_mobilenet.input, outputs=output)

# Tune MobileNet
mobilenet_model = create_mobilenet_model()
mobilenet_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                        loss='binary_crossentropy', metrics=['accuracy', 'Precision', 'Recall'])
print("\nTuning MobileNet (lr=0.0001, epochs=10)...")
mobilenet_history_tuned = mobilenet_model.fit(train_dataset_mobilenet, validation_data=val_dataset_mobilenet,
                                              epochs=10, callbacks=[early_stopping])

```

```python
print("\nMobileNet Tuned Results:")
print(f"Val Accuracy: {max(mobilenet_history_tuned.history['val_accuracy']):.4f}")
print(f"Val Loss: {min(mobilenet_history_tuned.history['val_loss']):.4f}")
```

```python
def get_predictions(model, dataset):
    preds = []
    true_labels = []
    for images, labels in dataset:
        preds.extend(model.predict(images, verbose=0).flatten())
        true_labels.extend(labels.numpy())
    return np.array(preds), np.array(true_labels)
```

```python
resnet_preds, val_labels = get_predictions(resnet_model, val_dataset_resnet)
mobilenet_preds, _ = get_predictions(mobilenet_model, val_dataset_mobilenet)
ensemble_preds = (resnet_preds + mobilenet_preds) / 2


ensemble_binary = (ensemble_preds > 0.5).astype(int)
```

```python

accuracy = accuracy_score(val_labels, ensemble_binary)
precision = precision_score(val_labels, ensemble_binary)
recall = recall_score(val_labels, ensemble_binary)
fpr, tpr, _ = roc_curve(val_labels, ensemble_preds)
roc_auc = auc(fpr, tpr)
```

```python
print("\nFinal Ensemble Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"AUC: {roc_auc:.4f}")
```
